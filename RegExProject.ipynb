```python
import re
import pandas as pd
```

```python
with open('Data.txt', 'r') as f:
    contents = f.read()
```

```python
print(contents)
```

```python
type(contents)
```

# matching Emails

```python
pattern = re.compile(r'[^@\s]+@[^@\s]+\.[^@\s]+')
matches = pattern.finditer(contents)

mails = pattern.findall(contents)

for match in matches:
    print(match)
```

# Matching phone numbers

```python
pattern = re.compile(r'\d{3}[-.]\d{3}[-.]\d{4}')
matches = pattern.finditer(contents)

phones = pattern.findall(contents)

for match in matches:
    print(match)
```
